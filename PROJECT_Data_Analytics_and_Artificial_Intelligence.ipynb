```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
data = {
    "ticker": ["AHH"] * 14,
    "open": [11.5, 11.66, 11.55, 11.63, 11.6, 11.6, 11.62, 11.7, 11.76, 11.76, 11.84, 11.64, 11.57, 11.73],
    "close": [11.58, 11.55, 11.6, 11.65, 11.53, 11.6, 11.74, 11.76, 11.73, 11.83, 11.75, 11.61, 11.7, 11.58],
    "adj_close": [8.49, 8.47, 8.51, 8.54, 8.45, 8.51, 8.61, 8.63, 8.6, 8.68, 8.62, 8.52, 8.58, 8.49],
    "low": [11.25, 11.5, 11.5, 11.55, 11.5, 11.54, 11.54, 11.7, 11.63, 11.61, 11.68, 11.54, 11.55, 11.57],
    "high": [11.68, 11.66, 11.6, 11.65, 11.6, 11.6, 11.74, 11.85, 11.83, 11.84, 11.84, 11.79, 11.73, 11.78],
    "volume": [4633900, 275800, 277100, 147400, 184100, 76800, 170300, 305400, 46800, 77000, 90200, 75400, 63400, 212400],
    "date": pd.date_range(start="2013-05-08", periods=14, freq="B")
}
df = pd.DataFrame(data)

# Clean
df["date"] = pd.to_datetime(df["date"])
df["decade"] = (df["date"].dt.year // 10) * 10
df.set_index("date", inplace=True)

# Summary
print(df.describe())

# Monthly Average Close
monthly_avg = df["close"].resample("M").mean()
monthly_avg.plot(title="Average Monthly Close Price")
plt.ylabel("Close Price")
plt.grid()
plt.show()

# Volume Histogram
sns.histplot(df["volume"], bins=10, kde=True)
plt.title("Volume Distribution")
plt.show()

# High and Low Boxplots
sns.boxplot(x=df["decade"], y=df["high"])
plt.title("Boxplot of High Prices by Decade")
plt.show()

sns.boxplot(x=df["decade"], y=df["low"])
plt.title("Boxplot of Low Prices by Decade")
plt.show()
```
